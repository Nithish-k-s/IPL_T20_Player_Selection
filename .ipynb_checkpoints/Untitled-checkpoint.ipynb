```python
import pandas as pd
import json
```

***(1)process match result***

```python
with open(r'C:\Users\Jyopc\Desktop\Jupyter\T20\t20_json_files\t20_wc_match_results.json') as f:
    data = json.load(f)
df_match = pd.DataFrame(data[0]['matchSummary'])
df_match.head()
```

```python
#how many elements in table?

df_match.shape
```

```python
#use score card as primary key
#so change the name to match_id
df_match.rename({'scorecard':'match_id'},axis = 1, inplace = True)
df_match.head()
```

**(2)process batting summary**

```python
#here there are many subsets length of data is total number of match so we need to include all subset 
with open(r"C:\Users\Jyopc\Desktop\Jupyter\T20\t20_json_files\t20_wc_batting_summary.json") as f:
    data = json.load(f)
    
    all_records = []
    
    for rec in data:
        all_records.extend(rec['battingSummary'])
        
df_batting = pd.DataFrame(all_records)
df_batting.head()
```

```python
#create a new column to print wheather a player is out or notout?
df_batting["out/not_out"] = df_batting.dismissal.apply(lambda x: "out" if len(x) > 0 else "not_out")
df_batting.head(11)
```

```python
#drop the dismisal column which is not needed?
df_batting.drop(columns=["dismissal"], inplace=True)
df_batting.head(11)
```

```python
#remove special characters in name?
df_batting['batsmanName'] = df_batting['batsmanName'].apply(lambda x: x.replace('â€',''))
df_batting.head(30)
```

```python
# we need to cnnect these two tables on match and matchid but we didnt have colums so create a dictionary having 
# team1 vs team2 : match_id 
match_id_dict = {}

for index, row in df_match.iterrows():
    key1 = row['team1'] + ' Vs ' + row['team2']
    key2 = row['team2'] + ' Vs ' + row['team1']
    
    match_id_dict[key1] = row["match_id"]
    match_id_dict[key2] = row["match_id"]
    
```

```python
#map match_id_dict with batting by the dictionary created so we can get a connection
df_batting["match_id"] = df_batting["match"].map(match_id_dict)
df_batting.head()
```

```python
#export match summary
df_match.to_csv('C:/Users/Jyopc/Desktop/Jupyter/T20/t20_csv_files/temp/dim_match_summary.csv', index=False)
```

```python
#export batting summary as csv file
df_batting.to_csv('C:/Users/Jyopc/Desktop/Jupyter/T20/t20_csv_files/temp/fact_batting_summary.csv', index=False)
```
